Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = '''SELECT WorkWeekHrs,CodeRevHrs,Age FROM master where Age between 30 and 35'''
df2 = pd.read_sql_query(QUERY, conn)
df_stacked = df2[(df2['Age']>29) & (df2['Age']<36)]
grouped = df_stacked.groupby(by='Age')
groupedm=grouped.median()
groupedm.plot(kind='bar', stacked=True)
plt.title('median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hrs')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = '''SELECT ConvertedComp, Age FROM master Where Age between 45 and 60 '''
df3=pd.read_sql_query(QUERY, conn)
df3_stacked = df3[(df3['Age']>44) & (df3['Age']<61)]
grouped = df3_stacked.groupby(by='Age')
groupedMedian=grouped.median()

groupedMedian.plot(kind='line')
plt.title('median `ConvertedComp` for all ages from 45 to 60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
```
